```python
import pandas as pd
import numpy as np
import spacy

nlp = spacy.load("en_core_web_lg")
```

```python
df = pd.read_excel('test.xlsx')
segmentation_df = df.copy()
df
```

```python
def lemaTag(text):
  postag_list = []
  doc = nlp(text)
  for token in doc:
    postag_list.append(token.lemma_)
  return postag_list
def posTag(text):
  postag_list = []
  doc = nlp(text)
  for token in doc:
    postag_list.append(token.pos_)
  return postag_list
def Tag(text):
  postag_list = []
  doc = nlp(text)
  for token in doc:
    postag_list.append(token.tag_)
  return postag_list
def stopTag(text):
  postag_list = []
  doc = nlp(text)
  for token in doc:
    postag_list.append(token.is_stop)
  return postag_list


segmentation_df['lemma_tag'] = df['Search'].apply(lambda x: lemaTag(x))
segmentation_df['lemma_tag'] = segmentation_df['lemma_tag'].apply(
    lambda y: np.nan if len(y) == 0 else ' '.join(y))

segmentation_df['pos_tag'] = df['Search'].apply(lambda x: posTag(x))
segmentation_df['tag'] = df['Search'].apply(lambda x: Tag(x))
segmentation_df['stop_tag'] = df['Search'].apply(lambda x: stopTag(x))


# for token in doc:
#     print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
#           token.shape_, token.is_alpha, token.is_stop)

segmentation_df
```

```python
# What is a user searching for?
def item(text):
  item_name = []
  doc = nlp(text)
  
  for token in doc:
    if token.pos == 92:
      item_name.append(str(token))
    
  
  if len(item_name) == 0:
    for token in doc.ents:
      if token.label == 383:
        item_name.append(str(token))
        print(token.text, token.label_)
        break

  if len(item_name) == 0:
    d = nlp(str(token.lemma_))
    item_name.append(str(d))
    

  return item_name


df['items'] = df.Search.apply(lambda x: item(x))
df['items'] = df['items'].apply(lambda y: np.nan if len(y) == 0 else ' '.join(y))

df


```

```python
# Where is he/she searching?
main_place = []

def location(text):
  place = []
  doc = nlp(text)
  location_name = False
  for token in doc:
    if location_name and token.pos == 96 or token.pos == 93:
      place.append(str(token))

    if token.pos == 85:
      location_name = True
    # print(token, token.pos_, token.pos)
  main_place.append(place)
  return place


df['search_location'] = df.Search.apply(lambda x: location(x))
df.search_location = df.search_location.apply(
    lambda y: '' if len(y) == 0 else ' '.join(y))
df

```

```python
# 3) What kind of products/services/things users are searching for? (Cheap, expensive, affordable, good, tasty, used, etc)
# noun pehlanu adj ke verb
def item_type(text):
  item = []
  doc = nlp(text)

  for token in doc:
    if token.pos == 84: #84
        item.append(str(token))
    if token.tag == 3822385049556375858:
        item.append(str(token))

  return item


df['item_type'] = df.Search.apply(lambda x: item_type(x))
df['item_type'] = df['item_type'].apply(
    lambda y: '' if len(y) == 0 else ' '.join(y))
df
```

```python
def get_props(text):
  props = []
  doc = nlp(text)
  for token in doc:
    if token.tag == 1292078113972184607:
      props.append(str(token))
  return props


df['preposition'] = df.Search.apply(lambda x: get_props(x))
df.preposition = df.preposition.apply(
    lambda y: np.nan if len(y) == 0 else ' '.join(y))
df
```

```python
df['users_location'] = df[(df['preposition'] == 'near')].search_location
df['users_location'] = df['users_location'].fillna('home')
df
```
